#### This is the code for interactiong GPT4, we need to use different prompts for different types of tasks. Here we show the prompt for Syntax Error Task(Only yes/No).

```python
#runs well with task 1(yes/no)
import csv
import openai

# Set your OpenAI API key
openai.api_key = 'Your Key'  # Be sure to replace this with your actual API key

# Function to read data from a CSV file into a list of dictionaries
def read_data(file_path):
    with open(file_path, mode='r', encoding='utf-8') as file:
        return list(csv.DictReader(file))

# Function to write data to a CSV file from a list of dictionaries
def write_data(file_path, data, fieldnames):
    with open(file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

# Path to your existing file and the new file
existing_file_path = "C:\\Users\\anany\\OneDrive\\Desktop\\sqlshare_mod_query_f.csv"
new_file_path = "C:\\Users\\anany\\OneDrive\\Desktop\\resultsqlshare_4(yesno1).csv"

# Read the existing data
data = read_data(existing_file_path)

# Iterate over each row in the data and call the OpenAI API to get the response
for row in data:
    sql_query = row.get("Modified_Statements", "")  # Replace "SQL Query Column" with the actual column name
    if sql_query:  # Check if the SQL query exists in the row
        conversation = [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"does this query syntactically correct: '{sql_query}'? Please answer with just YES or NO."},
        ]

        try:
            response = openai.ChatCompletion.create(
                model="gpt-4",
                messages=conversation,
                max_tokens=1,
                temperature=0,
                top_p=1
            )
            # Extract response and normalize to YES or NO
            api_response = response.choices[0].message['content'].strip().upper()
            if "YES" in api_response:
                row['Query Correct'] = 'YES'
            elif "NO" in api_response:
                row['Query Correct'] = 'NO'
            else:
                row['Query Correct'] = 'UNKNOWN'  # In case the response isn't clear
        except Exception as e:
            row['Query Correct'] = f"Error: {e}"
    else:
        row['Query Correct'] = 'NO QUERY'  # Mark rows without a query

# Write the updated data back to a new CSV file
fieldnames = list(data[0].keys()) + ['Query Correct']  # Adding 'Query Correct' column
write_data(new_file_path, data, fieldnames)

print(f"Updated file has been saved as '{new_file_path}'")
```

#### This is the code for interactiong GPT3.5, we need to use different prompts for different types of tasks. Here we show the prompts for Missing Type Task for Missing Token Task.

```python
#runs well in Task 2 'missing Type'
import csv
import openai

# Set your OpenAI API key
openai.api_key = 'Your Key'  # Be sure to replace this with your actual API key

# Function to read data from a CSV file into a list of dictionaries
def read_data(file_path):
    with open(file_path, mode='r', encoding='utf-8') as file:
        return list(csv.DictReader(file))

# Function to write data to a CSV file from a list of dictionaries
def write_data(file_path, data, fieldnames):
    with open(file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

# Path to your existing file and the new file
existing_file_path = "C:\\Users\\anany\\OneDrive\\Desktop\\sqlshare_mod_query_f.csv"
new_file_path = "C:\\Users\\anany\\OneDrive\\Desktop\\resultsqlshareMissingType.csv"

# Read the existing data
data = read_data(existing_file_path)

for row in data:
    sql_query = row.get("Modified Statements", "")  # Ensure this matches your actual column name
    if sql_query:
        conversation = [
            {"role": "system", "content": "You are a helpful assistant. When analyzing SQL queries, respond only with the type of missing component, such as 'Keyword', 'Table', 'Column', 'Alias', 'Value', 'Function', or 'Comparison'. If there's no error, respond with 'N/A'."},
            {"role": "user", "content": f"Analyze this SQL query and identify any missing components. SQL: '{sql_query}'"},
        ]

        try:
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",
                messages=conversation,
                max_tokens=10,  # Allows for a more complete response
                temperature=0.3,  # A lower temperature for more deterministic responses
                top_p=1
            )
            api_response = response.choices[0].message['content'].strip()

            # Explicitly checking for each expected response type
            if "Keyword" in api_response:
                row['Error Type'] = 'Missing Keyword'
            elif "Table" in api_response:
                row['Error Type'] = 'Missing Table'
            elif "Column" in api_response:
                row['Error Type'] = 'Missing Column'
            elif "Alias" in api_response:
                row['Error Type'] = 'Missing Alias'
            elif "Value" in api_response:
                row['Error Type'] = 'Missing Value'
            elif "Function" in api_response:
                row['Error Type'] = 'Missing Function'
            elif "Comparison" in api_response:
                row['Error Type'] = 'Missing Comparison'
            elif "N/A" in api_response:
                row['Error Type'] = 'N/A'
            else:
                # Default to a generic error type if the response doesn't match expected categories
                row['Error Type'] = 'Generic Error'
        except Exception as e:
            row['Error Type'] = f"Error: {e}"
    else:
        row['Error Type'] = 'No Query Provided'

# Proceed with writing the data to a new CSV file as before...
# Write the updated data back to a new CSV file
fieldnames = list(data[0].keys()) + ['Error Type']  # Adding 'Error Type' column
write_data(new_file_path, data, fieldnames)

print(f"Updated file has been saved as '{new_file_path}'")
```

```python
#llama3 Yes/No
import csv
import replicate

# Ensure the replicate library is installed: pip install replicate
# Set your Replicate API token directly (consider using environment variables for better security)
replicate.api_token = 'Your Key'  # Replace with your actual Replicate API token

# Function to read data from a CSV file into a list of dictionaries
def read_data(file_path):
    with open(file_path, mode='r', encoding='utf-8') as file:
        return list(csv.DictReader(file))

# Function to write data to a CSV file from a list of dictionaries
def write_data(file_path, data, fieldnames):
    with open(file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

# Path to your existing file and the new file
existing_file_path = "C:\\Users\\anany\\OneDrive\\Desktop\\sdss_mod_query_f.csv"
new_file_path = "C:\\Users\\anany\\OneDrive\\Desktop\\sdssresultllama2(yesno).csv"

# Read the existing data
data = read_data(existing_file_path)

# Iterate over each row in the data and call the Replicate API to get the response
for row in data:
    sql_query = row.get("Modified Statements", "")  # Replace "Modified Statements" with your actual column name if different
    if sql_query:  # Check if the SQL query exists in the row
        input = {
            "prompt": f"Is this SQL query syntactically correct: '{sql_query}'? Please answer with just YES or NO.",
            "temperature": 0.3,
            "max_new_tokens": 2
        }

        try:
            # Call the LLaMA 2 API with the prepared input
            output = replicate.run("meta/llama-2-70b-chat", input=input)
            api_response = "".join(output).strip().upper()
            
            # Normalize response to YES or NO
            if "YES" in api_response:
                row['Query Correct'] = 'YES'
            elif "NO" in api_response:
                row['Query Correct'] = 'NO'
            else:
                row['Query Correct'] = 'UNKNOWN'  # In case the response isn't clear
        except Exception as e:
            row['Query Correct'] = f"Error: {e}"
    else:
        row['Query Correct'] = 'NO QUERY'  # Mark rows without a query

# Write the updated data back to a new CSV file
fieldnames = list(data[0].keys()) + ['Query Correct']  # Adding 'Query Correct' column
write_data(new_file_path, data, fieldnames)

print(f"Updated file has been saved as '{new_file_path}'")
```
